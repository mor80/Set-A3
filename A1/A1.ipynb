```python
import pandas as pd
import matplotlib.pyplot as plt

filename = 'results.csv'
data = pd.read_csv(filename, sep=';', header=None, names=['n', 'approx_area'])

exact_area = 0.944517
data['relative_error'] = abs(data['approx_area'] - exact_area) / exact_area
```

Approximate area

```python
plt.figure(figsize=(10, 6))
plt.plot(data['n'], data['approx_area'], marker='o', linestyle='-', label='Approximate Area')
plt.axhline(y=exact_area, color='r', linestyle='--', label='Exact Area')
plt.xlabel('Number of Points (N)')
plt.ylabel('Approximate Area')
plt.title('Approximate Area')
plt.legend()
plt.savefig(f"approximate_area.png")
plt.grid()
plt.show()
```

Relative error

```python
plt.figure(figsize=(10, 6))
plt.plot(data['n'], data['relative_error'], marker='o', linestyle='-', label='Relative Error')
plt.xlabel('Number of Points (N)')
plt.ylabel('Relative Error')
plt.title('Relative Error')
plt.savefig(f"relative_error.png")
plt.legend()
plt.grid()
plt.show()
```
