```python
try:
    data_merge_sort = pd.read_csv("results1.csv", sep=';', header=None, names=["Size", "Time", "Type"])
    data_hybrid_sort = pd.read_csv("results2.csv", sep=';', header=None, names=["Size", "Time", "Type"])
except FileNotFoundError as e:
    print(f"Ошибка: {e}")
    raise

types = ["random", "reversed", "nearly_sorted"]
```

Sorting time random

```python
array_type = "random"
merge_data = data_merge_sort[data_merge_sort["Type"] == array_type]
hybrid_data = data_hybrid_sort[data_hybrid_sort["Type"] == array_type]

plt.figure(figsize=(10, 6))
plt.plot(merge_data["Size"], merge_data["Time"], marker='o', label="MergeSort")
plt.plot(hybrid_data["Size"], hybrid_data["Time"], marker='x', label="HybridMergeSort")
plt.title(f"Sorting Time for {array_type.capitalize()} Arrays")
plt.xlabel("Array Size")
plt.ylabel("Time (microseconds)")
plt.grid(True)
plt.legend()
plt.savefig(f"sorting_time_{array_type}.png")
plt.show()
```
